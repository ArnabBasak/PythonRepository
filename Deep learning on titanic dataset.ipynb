<b> Deep learning on titanic dataset </b>

```python
import pandas as pd
import numpy as np
```

```python
#import training data
train_set = pd.read_csv('titanic_train.csv')
```

```python
train_set.info()
```

```python
train_set.head(5)
```

```python
#splitting the data
y= train_set[['Survived']]
X = train_set[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
#converting the numpy array
X = X.astype(np.float32).values
y = y.astype(np.float32).values
```

```python
#importing the test data
test_set = pd.read_csv('titanic_test.csv')
```

```python
#cleaning and testing data
test_set = test_set.replace(['male','female'],[0,1])
test_set = test_set.replace(['S','C','Q'],[0,1,2])
test_set = test_set.fillna(0)
```

```python
X_test = test_set[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
X_test = X_test.astype(np.float32).values
```

```python
#importing keras
from keras.models import Sequential
from keras.layers import Dense
from keras.models import Sequential,load_model
from keras.layers import Dense,Dropout,BatchNormalization,Activation
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
seed = 42
np.random.seed(seed)
```

```python
#Data split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.2)
```

```python
#Model
model = Sequential()

#input layer
model.add(Dense(8,input_shape=(8,)))
model.add(BatchNormalization())
model.add(Activation("relu"))
model.add(Dropout(0.4))

```

```python
# hidden Layer
model.add(Dense(8))
model.add(BatchNormalization())
model.add(Activation("sigmoid"))
model.add(Dropout(0.4))

model.add(Dense(4))
model.add(BatchNormalization())
model.add(Activation("sigmoid"))
model.add(Dropout(0.4))

model.add(Dense(2,activation="sigmoid"))
```

```python
#output layer
model.add(Dense(1,activation='linear'))
```

```python
#model compile for binary classification
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#learning 
model.fit(X,y,nb_epoch=300,batch_size=30)
```

```python
#float to [0,1]
predictions = np.round(model.predict(X_test))
```

```python
predictions = pd.DataFrame(predictions)
```

```python
#result
result = pd.concat([train_set[['PassengerId']],predictions],axis=1)
```

```python
result.to_csv("result.csv",index=False)
```

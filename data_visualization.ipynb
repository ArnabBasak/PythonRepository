```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [40,50,51,52,60,48,30]
```

```python
plt.xlabel('Day')
plt.ylabel('templerature')
plt.title('weather')
plt.plot(x,y,color='red',linewidth=5,linestyle='dotted') #alpha is visiblity
```

```python
plt.plot(x,y,'rD') #keywword argument if
```

```python
day = [1,2,3,4,5,6,7]
max_temp = [51,52,48,49,56,42,45]
min_temp = [30,29,20,10,35,22,18]
avg_temp = [45,46,28,30,45,25,33]
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.plot(day,max_temp,label="max temperature")
plt.plot(day,min_temp,label="min temperature")
plt.plot(day,avg_temp,label="average temperature")
plt.legend(loc="lower left",shadow=True,fontsize="small")
plt.grid()
```

```python
import numpy as np
company=['GOOGLE','AMAZON','MICROSOFT','APPLE']
revenue=[90,136,89,70]
profit=[40,2,34,12]
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US tech stock")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(ypos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.yticks(ypos,company)
plt.title("US tech stock")

plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(ypos+0.2,profit,label="profit")
plt.legend()
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('suger range')
plt.ylabel('total number of patietnts')
plt.title('Blood suger analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],
         rwidth=0.95,color=['g','r'],label=['men','women'],orientation='horizontal')
plt.legend()
```
